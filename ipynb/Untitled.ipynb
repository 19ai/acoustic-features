```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft, dct
%matplotlib inline

N = 512
F = 1000
A = 32768
Fs = 80_000_000.0/32/128
T = N/Fs
t = np.linspace(0.0, T, N)
y = A * np.sin(2*np.pi*F*t)
plt.figure(figsize=[14,4])
plt.plot(t, y)
```

```python
plt.plot(np.abs(fft(y))/N)
```
