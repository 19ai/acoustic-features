# log10 fast approximation

Reference:
- https://community.arm.com/tools/f/discussions/4292/cmsis-dsp-new-functionality-proposal
- http://www.mathcentre.ac.uk/resources/Algebra%20leaflets/mc-logs3-2009-1.pdf
- https://www.ebayinc.com/stories/blogs/tech/fast-approximate-logarithms-part-i-the-basics/

$$ \log_{10}(x) = \log_{10}(2) \times \log_2(x) $$

```python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

logten2 = np.log10(2)

c = [1.23149591368684, -4.11852516267426, 6.02197014179219, -3.13396450166353]
def log10_approx(x):
    f, e = math.frexp(x)
    return logten2 * (c[0]*f**3 + c[1]*f**2 + c[2]*f + c[3] + e)

log10_approx_ = np.vectorize(log10_approx)

m = np.linspace(0, 0xffffffff, 1024)

plt.figure(figsize=[12,6])
plt.plot(log10_approx_(m))
plt.plot(np.log10(m))
```
