## DCT via FFT

Reference: https://dsp.stackexchange.com/questions/2807/fast-cosine-transform-via-fft/10606

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft, dct
%matplotlib inline

N = 1024
t = np.linspace(0.0, 1.0, N)
y = (np.sin(2*np.pi*100*t)+np.sin(2*np.pi*250*t))/2
plt.figure(figsize=[14,4])
```

```python
# DCT
plt.figure(figsize=[14,4])
plt.plot(20 * np.log10(np.abs(dct(y))))
dct(y)[:12]
```

```python
# DCT via FFT
yy = np.zeros(2*N)
yy[N:] = y[::-1]
k = np.arange(N)
Y = fft(yy)[:N]
#Y = Y * 2 * np.exp(-1j*np.pi*k/(2*N))
Y = Y * 2 * (np.cos(-1*np.pi*k/(2*N)) + 1j * np.sin(-1*np.pi*k/(2*N)))
Y = Y.real
plt.figure(figsize=[14,4])
plt.plot(20*np.log10(np.abs(Y)))
```

## Mel scale

```python
f = np.linspace(0, 22000, N)
```

```python
logten2 = np.log10(2)
import math

c = [1.23149591368684, -4.11852516267426, 6.02197014179219, -3.13396450166353]
def log10_approx(x):
    f, e = math.frexp(x)
    return logten2 * (c[0]*f**3 + c[1]*f**2 + c[2]*f + c[3] + e)

log10_approx_ = np.vectorize(log10_approx)

plt.figure(figsize=[14,4])
plt.plot(2595 * log10_approx_(1 + f/700))
plt.plot(2595 * np.log10(1 + f/700))
```
